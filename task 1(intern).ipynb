```python
#Filters
df1 = df[
    (df['Rating'] >= 4.0) &
    (df['Size_MB'] >= 10) &
    (df['Last_Updated'].dt.month == 1)
]

#Top 10 Categories by Installs
top10 = df1.groupby('Category')['Installs'].sum().nlargest(10).index
df1 = df1[df1['Category'].isin(top10)]

#Aggregation
agg = df1.groupby('Category').agg({
    'Rating': 'mean',
    'Reviews': 'sum'
}).reset_index()

#Plotly Chart
import plotly.graph_objects as go

fig = go.Figure([
    go.Bar(x=agg['Category'], y=agg['Rating'], name='Avg Rating'),
    go.Bar(x=agg['Category'], y=agg['Reviews'], name='Total Reviews')
])

fig.update_layout(barmode='group')
```
